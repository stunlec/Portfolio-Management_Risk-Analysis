# Solution 1

```python
import pandas as pd
import numpy as np
from math import pow,sqrt
```

```python
def moment_dist(ser,deg):
    sum_val = ser.sum()
    mean = sum_val/len(ser)
    if deg==1:
        return mean
    
    sum_diff = 0
    for i in range(len(ser)):
        diff = ser[i]-mean
        sum_diff+=pow(diff,2)
        
    var = sum_diff/(len(ser)-1)
    SD = sqrt(var)
    
    if deg==2:
        return var
    
    sum_diff2 = 0
    if deg==3:
        for i in range(len(ser)):
            diff2 = ser[i]-mean
            num = diff2/SD
            sum_diff2 += pow(num,3)
    
        skew = sum_diff2/len(ser)
        return float(skew)
    
    if deg==4:
        for i in range(len(ser)):
            diff3 = ser[i]-mean
            num1 = diff3/SD
            sum_diff2 += pow(num1,4)
    
        kurt = sum_diff2/len(ser)
        return float(kurt)
```

# Solution 2

```python
def Jarque_bera(ser):
    kurt = float(moment_dist(ser,4))
    skew = float(moment_dist(ser,3))
    n = float(len(ser))
    
    JB = (n/6)*((skew**2) + (1/4)*(kurt-3.0)**2)
    
    if JB>=0.0 and JB<=0.05:
        return True
    else:
        return False
```

No, the real life stocks do not closely follow the normal distribution. For real life stocks to follow normal distribution its standard deviation has to be zero and kurtosis to be three. Jarque Bera(JB) test is used to see whether a distribution follows a normal distribution. If the JB is 0 or less than 0.05 we consider the distribution to be normal. We can test stocks from the csv whether they follow normal distribution or not ny using function in solution 2.

```python
file=pd.read_csv(r"edhec-hedgefundindices.csv")
```

```python
df=pd.DataFrame(file)
abcd = df['CTA Global']
for i in range(len(abcd)):
    print(abcd[i])
moment_dist(abcd,1)
```

```python
column_list = list(df.columns.values)
column_list.remove("date")

for i in column_list:
    print("The stock",i,"follows Normal Distribution: ",Jarque_bera(df[i]))
```
