```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
def ann_ror(l_stock):
    n=len(l_stock)
    eyp=float(l_stock[n-1])
    byp=float(l_stock[0])
    ror=((eyp-byp)/byp)*100
    return ror
```

```python
file=pd.read_csv(r"C:\Users\shash\Downloads\sample_prices.csv")
df = pd.DataFrame(file)
blue = df['BLUE']
def gauss_var(ser,alpha=5):
    mean = np.mean(ser)
    sdev = np.std(ser)
    z = -1.65
    var = (mean+(z*sdev))
    return var

def gauss_cvar(ser,alpha=5):
    ser = np.array(ser)
    var = gauss_var(ser)
    sum_ret = 0
    for i in range(len(ser)):
        if(ser[i]<var):
            sum_ret+=ser[i]
    cvar = sum_ret/len(ser)
    return cvar
print(blue)
gauss_cvar(blue)
blue.plot()
```
